## Export database to csv

One way to **dump a database in sqlite3 to csv**. 


Simply as an exercise and a utility.

```python
import sqlite3
import csv
import pandas as pd

# Conectar a la base de datos SQLite
conn = sqlite3.connect('database/ksprojects.db')  # Reemplaza 'tudatabase.db' con el nombre de tu base de datos SQLite
cursor = conn.cursor()

# Ejecutar una consulta SQL para seleccionar datos de la tabla
cursor.execute("SELECT * FROM ksprojects")  # Reemplaza 'nombre_de_la_tabla' con el nombre de tu tabla

# Obtener los resultados de la consulta
rows = cursor.fetchall()

# Cerrar la conexión a la base de datos
conn.close()

# Escribir los datos en un archivo CSV
with open('ksprojects_datos_exportados.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    
    # Escribir el encabezado (nombres de las columnas)
    column_names = [description[0] for description in cursor.description]
    csv_writer.writerow(column_names)
    
    # Escribir los datos
    csv_writer.writerows(rows)

print("Exportación completada. Los datos se han guardado en 'ksprojects_datos_exportados.csv'.")
```
