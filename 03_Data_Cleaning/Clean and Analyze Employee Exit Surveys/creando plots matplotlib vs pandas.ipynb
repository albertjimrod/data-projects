```python
import matplotlib.pyplot as plt
```

```python
x_coordinates_1 = [6,5,1,6,1]
y_coordinates_1 = [22,2,14,2,81]

x_coordinates_2 = [2,5,1,6,8]
y_coordinates_2 = [5,15,5,5,81]

x_coordinates_3 = [2,1,1,6,8]
y_coordinates_3 = [0,0,88,8,81]
```

```python
plt.plot(x_coordinates_1, y_coordinates_1) # dibujando una linea.
plt.show()
```

```python
plt.title("Titulo")
plt.xlabel("x_label")
plt.ylabel("y_label")
```

### Dibujando tres gráficas que comparten el mismo eje.

Comparten el mismo eje, la cantidad de valores deben ser de la misma longitud.

```python
plt.plot(x_coordinates_1, y_coordinates_1) # dibujando una linea.
plt.plot(x_coordinates_2, y_coordinates_2)
plt.plot(x_coordinates_3, y_coordinates_3)
plt.legend("123") # cada uno de los caracteres es una de las graficas.
plt.show()
```

### Dibujando dos gráficas sin que compartan el mismo eje.

```python
plt.plot(x_coordinates_1, y_coordinates_1) # dibujando una linea.
plt.show()
plt.plot(x_coordinates_3, y_coordinates_3)
plt.legend("123") # cada uno de los caracteres es una de las graficas.
plt.show()
```

## Importing seaborn and activating the default style settings:

```python
import seaborn as sns
sns.set_theme()
```

### Plotting a relational plot with five variables:

```python
sns.relplot(data=df, x='seen', y='synt_brand',
           hue='city', palette= 'RdYlGn',
           size='price' , sizes=(1,300),
           style='city')
plt.show()
```

```python
df.head(5)
```

```python
plt.figure(figsize=(10,12))
for i in range(1,5):
    plt.subplot(2,2,i)
    #plt.plot(x,y)
    df['sell'].plot(kind='line')
```

### Separating a relational plot based on a sixth variable:

```python
sns.relplot(data=df, x='seen', y='synt_brand',
           hue='city', palette= 'RdYlGn',
           size='price' , sizes=(1,300),
           style='city', col='synt_brand')
plt.show()
```
